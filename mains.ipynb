```python
import os
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
from fastapi import FastAPI
import service as s
```

```python
def dataFetch():
    df = pd.read_csv('Approved_Ideas.csv', encoding= 'unicode_escape')
    #df1 = pd.read_csv('List_of_Not_Approved_Idea.csv', encoding= 'unicode_escape')
    #df.add(df1)
    return df

def modelFetch():
    model = SentenceTransformer('paraphrase-albert-small-v2')
    return model
```

```python
app = FastAPI()
@app.get("/innovationDuplication")
def root(inputS):
        df = dataFetch()
        model = modelFetch()
        data = ['India is my country','Hello World','Example','who are you']
        #data = df['Briefly provide the market data for the potential idea/ innovation'].to_list()
        data.append(inputS)
        sentence_vec = model.encode(data)
        i = len(data)-1
        #Comparing First Innovation Idea with remaining other 10 Innovation Ideas
        compare = cosine_similarity(
                            [sentence_vec[i]],#Consider 1st Sentence 
                            sentence_vec[0:i+1]) #Compare with remaining sentence
        li = []
        k=0
        for k in range(compare.size-1) :
            if((compare[0][k]*100)>85.00):
                        #li.append(k)
                        #outputS = inputS + " :::::::::::::::::::::::::::::: is ################   " + str(compare[0][k]*100) + "   ############## similar to ::::::::::::::::::::::::::::  " + " " + data[k]
                        outputS = { "Idea_Given": inputS, "Old_Idea": data[k], "Percentage_Comparison": compare[0][k]*100}
                        li.append(outputS)
        if not li :
            outputS = "Congratz!!  New Idea"
            li.append(outputS)    
        # Item.Idea_Given = inputS
        # Item.Old_Idea = data[k]
        # Item.Percentage_Comparison = compare[0][k]*100

        #return li
        #return Item
        #response = { "Idea_Given": inputS, "Old_Idea": data[k], "Percentage_Comparison": compare[0][k]*100}
        return outputS
```
